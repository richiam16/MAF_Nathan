# Code to replicate the Glycine Content, Hydrophilicity Averages, IDRs Ratio and The Figure as Shown in the Publication 

```python
import pandas as pd
import numpy as np
from Bio.SeqUtils.ProtParam import ProteinAnalysis
import seaborn as sns
import matplotlib.pyplot as plt
from Bio import SeqIO
from Bio.SeqUtils.ProtParam import ProteinAnalysis
import seaborn as sns
import matplotlib.pyplot as plt
```

## Data Manipulation from Mtuberculosis UniProt Tables and the Hits Table

```python
# Read the excel-file file from mtuberculosis proteins

mtubercolisis_df = pd.read_excel("mtuberculosis.xlsx")
```

```python
# Read the NOCSD file wit Uniprot information

noscd_df = pd.read_csv("counts_all_stages_MAGECK_alphafold.csv")
# I need to create a clean version of this file
```

```python
# Change the name of one the coloumns to make the merge
noscd_df = noscd_df.rename(columns={"Uniprot_entry": "Entry"})
```

```python
# Combine the dataframe

master_df = pd.merge(mtubercolisis_df, noscd_df[["Counts_1st", "Counts_2nd", "Counts_3rd", "Counts_4th", "Total_Counts", "Entry","id"]], on='Entry', how='right')
```

## Calculation of the Ratio of IDRs based on the SETH Model (PMCID: PMC9580958)

```python
# Make a FASTA file to use with the colab version of SETH (https://colab.research.google.com/drive/1vDWh5YI_BPxQg0ku6CxKtSXEJ25u2wSq?usp=sharing)

def df_to_fasta(df, file_name):
    with open(file_name, 'w') as f:
        for index, row in df.iterrows():
            f.write(f">{row['Entry']}\n")
            f.write(f"{row['Sequence']}\n")

df_to_fasta(master_df, 'proteins_hits.fasta')
```

```python
# Collect the information from the FASTA file that is the result of running the colab of the SETH model
ids = []
sequences = []
descriptions = []

# Read the FASTA file
for record in SeqIO.parse("disorder_preds.fasta", "fasta"):
    ids.append(record.id)
    sequences.append(str(record.seq))
    descriptions.append(record.description)
```

```python
# Converting the File into a dataframe to merge

df_predictions = pd.DataFrame({
    'Entry': ids,
    'SETH_predictions': sequences
})
```

```python
# Merging the dataframe of the predictions, the predictions will be in the sequence 

master_df = pd.merge(master_df, df_predictions, on='Entry', how='outer')
```

```python
# SETH is based on the CheZOD scores, where a residue value of CheZOD score below 8 represent IDRs
# Counting the number of IDRs in all the proteins

def count_(row):
    try:
        values = [float(a) for a in row["SETH_predictions"].split(",")]
        number_irdf = sum(1 for value in values if value < 8.0)
        return number_irdf
    except AttributeError:
        return np.nan

master_df["number_disorder_residues"] = master_df.apply(count_, axis=1)
```

```python
# Calculation of the Ratio of IDRs in each protein

master_df["Ratio_IDRs"] = master_df["number_disorder_residues"]/master_df["Length"]
```

## Calculation of the Hydrophilicity Average Using Biopython

```python
# Calculate the Hydrophilicity Average

# Parker Hydrophilicity scale derived from HPLC peptide retention times
# Biochemistry 25:5425-5431(1986)
# "Parker"
# Values obtained from: https://github.com/biopython/biopython/blob/master/Bio/SeqUtils/ProtParamData.py

pa = {"A": 2.1, "R": 4.2, "N": 7, "D": 10, "C": 1.4,
      "Q": 6, "E": 7.8, "G": 5.7, "H": 2.1, "I": -8,
      "L": -9.2, "K": 5.7, "M": -4.2, "F": -9.2, "P": 2.1,
      "S": 6.5, "T": 5.2, "W": -10, "Y": -1.9, "V": -3.7}

def calculate_hydro_index(row):
    pa = {"A": 2.1, "R": 4.2, "N": 7, "D": 10, "C": 1.4,
      "Q": 6, "E": 7.8, "G": 5.7, "H": 2.1, "I": -8,
      "L": -9.2, "K": 5.7, "M": -4.2, "F": -9.2, "P": 2.1,
      "S": 6.5, "T": 5.2, "W": -10, "Y": -1.9, "V": -3.7}
    try:
        X = ProteinAnalysis(row["Sequence"])
        a = X.protein_scale(pa, 2)
        h_index = sum(a) / len(a)
        return h_index
    except AttributeError:
        return np.nan
    

master_df["Hydrophilicity_Average"] = master_df.apply(calculate_hydro_index, axis=1)
```

## Calculation of Glycine Content Using Biopython

```python
# Calculate Glycine Content

def calcualte_amonoacid_content(row):
    try:
        X = ProteinAnalysis(row["Sequence"])
        return X.get_amino_acids_percent()["G"]*100
    except AttributeError:
        return np.nan
        

master_df["Glycine_content"] = master_df.apply(calcualte_amonoacid_content, axis=1)
```

## Plotting the Three Features in 3 dimensions

```python
# Plotting the three parameters
x = master_df["Glycine_content"]
y = master_df["Hydrophilicity_Average"]
z = master_df["Ratio_IDRs"]


# Adding customization
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(x, y, z, c=z, cmap='viridis', marker='o')
ax.set_xlabel('Glycine Content',fontsize=16)
ax.set_ylabel('Hydrophilicity Average', fontsize=16)
ax.set_zlabel("Ratio of IDRs", fontsize=16)

# Adding a colorbar
cbar = plt.colorbar(scatter, shrink=0.6)
cbar.set_label('Ratio of IDRs')
plt.savefig("3D_figure.png", dpi=300)
plt.show()
```

```python
# Printing the dataframe of the proteins that are in the cluster (Glycine content higher than 30 and Hydrophilicity average higher than 2.5)

top_50 = master_df[(master_df["Glycine_content"] > 30) & (master_df["Hydrophilicity_Average"] > 2.5)]

# Ordering by number of NOSCD

top_50 = top_50.sort_values(by="Counts_3rd", ascending=False)
```

```python
# Extra graphs of CheZOD score of the proteins in the cluster:

def make_plots_IDRs(row):
    values = [float(a) for a in row["SETH_predictions"].split(",")]
    index_ = [i for i in range(0,len(values))]
    plt.plot(index_, values)
    plt.plot(index_, [8 for i in range(0,len(values))])
    name, ratio = row["Gene Names"], row["Ratio_IDRs"]
    plt.title(f"{name}, Ratio of IDRs:  {ratio}")
    plt.xlabel("Residue Index")
    plt.ylabel("CheZOD score")
    #plt.savefig(f"/home/richiam/alphafold_project/test/dfs/idrs_plots/{entry}.jpg")
    plt.show()

top_50.apply(make_plots_IDRs, axis=1)
```
